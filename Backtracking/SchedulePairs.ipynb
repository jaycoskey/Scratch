```python
dates = list(set(dates.replace('	', ' ').split()))
dates = sorted(dates)
dates = dates[0:13]  # TODO: Extend list of dates
dates
```

```python
pairs = [(names[k1], names[k0])
            for k0 in range(len(names))
            for k1 in range(k0)
        ]
print(pairs)
```

```python
# Create schedule
assert(len(names) % 2 == 0)

def scheduled(day_pairs):
    print('INFO: scheduled: day_pairs={}'.format(day_pairs))
    result = []
    for p in day_pairs:
        result.append(p[0])
        result.append(p[1])
    return result

available_pairs = pairs.copy()

schedule = defaultdict(list)  # Keys: dates, Values: Lists of pairs
for date in dates:
    # print('**** INFO: date={} ****'.format(date))
    day_pairs = schedule[date]
    # print('Checkpoint hotel', flush=True)
    have_refreshed_pairs = False
    while len(schedule[date]) < len(names) / 2:
        # print('Checkpoint india', flush=True)
        
        # Find the next pair for the given date
        for p in available_pairs:
            # print('INFO: pair={}'.format(p))
            # Check to see if both people in the pair are available.  If not, move on.
            if (not p[0] in scheduled(day_pairs)) and (not p[1] in scheduled(day_pairs)):
                # print('== About to append pair ==')
                schedule[date].append(p)
                available_pairs.remove(p)
                break
            else:
                # print('Pair not available')
                pass
        print('Remaining pairs: {}'.format(available_pairs))
        
        # Refresh the list of available pairs at most once per date
        # if not have_refreshed_pairs and len(available_pairs) == 0:
        #    available_pairs = pairs.copy()
        #    have_refreshed_pairs = True
```
